# 1.LOAD LIBRARIES

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

print("Setup is working")
```

# 2.LOAD DATASET 

```python
# CELL 2 — Load Dataset
# We use seaborn's built-in iris dataset. 
# It has 150 rows and 5 columns: 4 features + 1 target (species)
df = sns.load_dataset('iris')
df.head()
```

# 3.LOAD INSPECTION

```python
# CELL 3 — Shape and Columns
# shape tells us rows x columns
# columns tells us feature names
print("Shape:", df.shape)
print("Columns:", df.columns.tolist())
```

```python
# CELL 4 — Data Types and Nulls
# .info() shows column types and if any values are missing
df.info()
# INSIGHT: All 4 features are float64, species is object (categorical). No nulls.
```

```python
# CELL 5 — Summary Statistics
# .describe() gives mean, std, min, max etc. for numeric columns
df.describe()
# INSIGHT: sepal_length ranges 4.3-7.9, petal_length has high variance (big spread)
```

```python
# CELL 6 — Scatter Plot
# PURPOSE: See if species are separable by sepal measurements
sns.scatterplot(x='sepal_length', y='sepal_width', hue='species', data=df)
plt.title("Sepal Length vs Sepal Width by Species")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Sepal Width (cm)")
plt.tight_layout()
plt.show()
# INSIGHT: Setosa is clearly separated. Versicolor and Virginica overlap slightly.
```

```python
# CELL 7 — Histograms
# PURPOSE: Understand distribution of each feature
df.hist(figsize=(10,8), color='steelblue', edgecolor='black')
plt.suptitle("Feature Distributions - Iris Dataset", y=1.02)
plt.tight_layout()
plt.show()
# INSIGHT: Petal length and width show bimodal distribution — suggests species differences
```

```python
# CELL 8 — Box Plots
# PURPOSE: Spot outliers and compare spread across features
sns.boxplot(data=df.drop(columns='species'))
plt.title("Feature Distribution and Outliers - Iris Dataset")
plt.xlabel("Features")
plt.ylabel("Value (cm)")
plt.tight_layout()
plt.show()
# INSIGHT: Sepal width has the most outliers. Petal features have less spread.
```
